# Get BSE data

End of day prices, and additional information for companies trading on the Bombay Stock Exchange in India.

```python
# Import modules
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import datetime
import csv
import requests
import pandas_datareader.data as web
import pandas_datareader as pdr
from pandas_datareader import data, wb
import quandl

%matplotlib inline
```

```python
# quandl API key
quandl_api_key = 'TOMATO'
quandl.ApiConfig.api_key = 'TOMATO'
```

```python
# Input Start and End Date
#start = datetime.datetime(2019,1,1)
end = datetime.datetime(2018,12,31).strftime ("%Y-%m-%d")

Today = datetime.datetime.now().strftime ("%Y-%m-%d")

# days_to_subtract = 7
# last_week = datetime.datetime.today() - timedelta(days=days_to_subtract)
# last_week = last_week.strftime ("%Y-%m-%d")
```

```python
# Import list of all avaliable stock names from Quandl meta-data file
link = 'https://www.quandl.com/api/v3/databases/BSE/metadata?api_key='+quandl_api_key

import urllib.request
urllib.request.urlretrieve(link, "BSE_metadata.csv.zip")

import zipfile
compressed_file = zipfile.ZipFile('BSE_metadata.csv.zip')
csv_file = compressed_file.open('BSE_metadata.csv')
BSE_metadata = pd.read_csv(csv_file)

# remove no longer active stocks
# BSE_metadata = BSE_metadata[BSE_metadata.from_date >= start]
BSE_metadata = BSE_metadata[BSE_metadata.to_date >= end]
```

```python
display(BSE_metadata.info())
display(BSE_metadata.head(3))
```

```python
# Get only the list of stock names - remove everything else
BSE_metadata['stock_name'] = "BSE/"+ BSE_metadata['code']
stock_list = BSE_metadata['stock_name'].tolist()
stock_list = list(set(stock_list))
stock_list = np.sort(stock_list)

print("Available stocks", len(stock_list), "\nSample:", stock_list[:3])
```

```python
%%time 
# Create Empty Dataframe
stock_final = pd.DataFrame()

# Scrape the stock prices from quandl
for i in range(len(stock_list)) :  
    #print(i)    
    try:
        stock=[]
        stock = web.DataReader(stock_list[i],"quandl",
                               BSE_metadata['from_date'][i],end,
                               access_key=quandl_api_key)
        stock['Name']=stock_list[i]
        stock_final = pd.DataFrame.append(stock_final,stock, sort=False)
        print("Added",i+1, BSE_metadata["BSE/"+BSE_metadata.code  == stock_list[i]]["name"].values)
    
    except Exception:
        print('Not found!',i+1, BSE_metadata["BSE/"+BSE_metadata.code  == stock_list[i]]["name"].values)
        i = i+1     
```

```python
stock_final.reset_index(inplace=True)
stock_final = stock_final.sort_values(['Name', 'Date'])

print("Available stocks", len(stock_list))
print("Downloaded stocks", stock_final.Name.nunique())
display(stock_final.info())
display(stock_final.sample(3))
```

```python
# Download the appended stock prices
stock_final.to_csv('data/BSE_till2018_'+Today+'.csv')
# add data from 1 Jan 2019 in new dataframe
```
